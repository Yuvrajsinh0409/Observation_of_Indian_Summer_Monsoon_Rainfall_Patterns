# Data Structures which Xarray provides

**Xarray Provides 2 types of data structures:-**

**1. DataArrays 2. Datasets**

**DataArrays are consists of multidimensional arrays just as a normal numpy array provides**

**But The XArray DataArrays provides more functionality by providing some attributes such as:-**


**(1) Variables (2) Dimensions (3) Coordinates (4) Attributes**

**We know that a 2D matrix has a column name , a row name , and data according to each cell (intersecting that row and column)**

**Now suppose we have to plot a graph according to that matrix so what we can do is , we can take the row as x-axis and column as y-axis and plot a graph wherever the values of rows and column is intersecting with eachother**

**The same way , here we are capable of providing our own custom row and column values and their names**

**In technical terms , these row's and column's can be refered as dimensions and their values can be refered as coordinates**

**The values inside matrix which each cell contains can be refered as variable's values , there can be multiple variables according to these coordinates , through which we can plot different types of graphs**

**Attributes are additional information on the variables values such as validity , type, unit , startdate , enddate etc...**

**Using these coordinates and dimensions we can provide that our graph's x and y lables are going to start from this lower to this higher value , and using variables we can provide that at this intersecting point intensity is this much high/low**
